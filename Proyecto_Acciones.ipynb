```python
# *****Vamos a revisar el cambio en precio de las acciones a lo largo del tiempo *****

# Podemos importar pandas y numpy para manejar y manipular datos de manera eficiente.
import pandas as pd
import numpy as np


# También estamos importando matplotlib y seaborn para la visualización de datos.
# Establecemos el estilo de fondo de seaborn como 'whitegrid' para un aspecto limpio.
# Además, aplicamos el estilo de gráficos de fivethirtyeight a nuestros gráficos.
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
# Usamos "%matplotlib inline" para que los gráficos se muestren directamente en la celda
%matplotlib inline

# Podemos importar DataReader y pdr (pandas_datareader) para obtener datos financieros de Yahoo
from pandas_datareader.data import DataReader
# Ahora que hemos instalado la biblioteca, podemos importarla en nuestro código.
import yfinance as yf
from pandas_datareader import data as pdr

# Configuramos yfinance para sobrescribir el pdr predeterminado y utilizar su funcionalidad mejorada
yf.pdr_override()

# Importamos el módulo datetime para trabajar con fechas y horas
from datetime import datetime


# Creamos una lista de símbolos de acciones tecnológicas que nos interesan
tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'NFLX', 'BTC']

# Definimos las fechas de inicio y fin para obtener datos durante el último año
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)

# Utilizamos un bucle for para descargar datos de acciones para cada símbolo en tech_list.
# Creamos variables globales con los nombres de los símbolos para facilitar su acceso más adelante.
for stock in tech_list:
    globals()[stock] = yf.download(stock, start, end)

# Creamos una lista de las variables de las empresas que hemos descargado
company_list = [AAPL, GOOG, MSFT, AMZN, NFLX, BTC]
# Asignamos nombres de empresas y criptomonedas correspondientes a cada conjunto de datos
company_name = ["APPLE", "GOOGLE", "MICROSOFT", "AMAZON", "NFLX", "BTC"]

# Utilizamos un bucle for para agregar una nueva columna 'company_name' a cada conjunto de datos
for company, com_name in zip(company_list, company_name):
    company["company_name"] = com_name

# Concatenamos los conjuntos de datos en un solo DataFrame llamado 'df'
df = pd.concat(company_list, axis=0)
# Mostramos las últimas 100 filas del DataFrame resultante
df.tail(100)
```

```python
# Seleccionamos los primeros 'training_data_len' datos escalados para el conjunto de entrenamiento
train_data = data.values[0:int(training_data_len), :]

# Inicializamos listas vacías para las características (x_train) y las etiquetas (y_train).
x_train = []
y_train = []

# Creamos ventanas temporales de datos para el entrenamiento.
for i in range(60, len(train_data)):
    # x_train contiene ventanas temporales de 60 pasos de tiempo para cada acción.
    x_train.append(train_data[i-60:i, :])
    # y_train contiene el siguiente valor después de la ventana temporal de 60 pasos para cada acción.
    y_train.append(train_data[i, :])
    # Si estamos en las primeras dos iteraciones, mostramos las ventanas y etiquetas
    if i <= 61:
        print(x_train)
        print(y_train)
        print()

# Convertimos las listas a arreglos NumPy
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape: Agregamos una dimensión adicional para representar el tiempo en las características
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], x_train.shape[2]))
```

```python
# Importamos las clases necesarias de Keras.
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Construimos el modelo LSTM
# Creamos un modelo secuencial de Keras
model = Sequential()

# Agregamos una capa LSTM con 128 unidades y que devuelve secuencias (return_sequences=True).
# La forma de entrada es (x_train.shape[1], 1) que representa el tiempo y las características.
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
# Agregamos otra capa LSTM con 64 unidades que no devuelve secuencias (return_sequences=False).
model.add(LSTM(64, return_sequences=False))
# Agregamos una capa densa con 25 unidades.
model.add(Dense(25))
# Agregamos una capa densa final con 1 unidad (predicción de un solo valor).
model.add(Dense(1))

# Compilamos el modelo utilizando el optimizador 'adam' y la función de pérdida 'mean_squared_error'.
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenamos el modelo con los datos de entrenamiento (x_train, y_train).
# Utilizamos un tamaño de lote de 1 y entrenamos durante 10 épocas.
model.fit(x_train, y_train, batch_size=1, epochs=10)
```
